```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
```

# Episode Transcripts

## 1. Rebirth 

```python
# Get the url
rebirth='https://deathnote.fandom.com/wiki/Rebirth/Transcript'
page=requests.get(rebirth)
page.encoding='utf-8'
# Beautify
soup=BeautifulSoup(page.text, 'html')

# Extract the column with the characters' names
actor =soup.find_all('th')
characters=[character.text.strip() for character in actor]

# Extract their dialogue
speech=soup.find_all('td')
dialogue=[dialog.text.strip() for dialog in speech]

# Assign the data in two different data frames
characters = pd.DataFrame({'Characters': characters})
dialogue = pd.DataFrame({'Dialogue': dialogue})

# Drop the first 2 rows in the dialogue data frame as they are irrelevant

dialogue=dialogue.drop([0,1]).reset_index(drop=True)

# Identify the indices of blank rows. These blsnk rows represent the narrators' character 
blank_indices = dialogue[dialogue['Dialogue'].str.strip() == ''].index
blank_indices
# Create a list of indices to drop (blank rows and the ones immediately after) We are dropping the rows after the blank rows as those represent
#the narrators dialogue

drop_indices = set(blank_indices)
drop_indices.update(blank_indices + 1)

# Drop the rows with these indices
dialogue = dialogue.drop(index=drop_indices)
dialogue = dialogue.reset_index(drop=True)

# Ensure both DataFrames have the same number of rows
if len(characters) != len(dialogue):
    raise ValueError("Both DataFrames must have the same number of rows")

# Insert the column from df2 into df1 at the second position
rebirth=characters.insert(1, 'Dialogue', dialogue['Dialogue'])

# Insert a column from of the Eoisode name at the beginning
characters.insert(0, 'Episode Name', 'Rebirth')
rebirth=characters
print(rebirth.head(1))
rebirth.tail(1)
```

## 7. Overcast

```python
# Get the url

overcast='https://deathnote.fandom.com/wiki/Overcast/Transcript'
page=requests.get(overcast)
page.encoding='utf-8'

# Beautify

soup=BeautifulSoup(page.text, 'html')

# Extract the column with the characters' names

actor =soup.find_all('th')
characters=[character.text.strip() for character in actor]

# Extract their dialogue

speech=soup.find_all('td')
dialogue=[dialog.text.strip() for dialog in speech]

# Assign the data in two different data frames

characters = pd.DataFrame({'Characters': characters})
dialogue = pd.DataFrame({'Dialogue': dialogue})

# Drop the first 2 rows in the dialogue data frame as they are irrelevant

dialogue=dialogue.drop(range(8)).reset_index(drop=True)

# Fixing the issue with dialogue.iloc[66] Matsuda, a character has been placed in the wrong data frame

# Create a new DataFrame for the extra row

new_row = pd.DataFrame({'Characters': ['Matsuda']})
characters_part1 = characters.iloc[:66]
characters_part2 = characters.iloc[66:]
characters = pd.concat([characters_part1, new_row, characters_part2]).reset_index(drop=True)
dialogue = dialogue.drop(66).reset_index(drop=True)

# Identify the indices of blank rows. These blsnk rows represent the narrators' character 

blank_indices = dialogue[dialogue['Dialogue'].str.strip() == ''].index
blank_indices

# Create a list of indices to drop (blank rows and the ones immediately after) We are dropping the rows after the blank rows as those represent
#the narrators dialogue

drop_indices = set(blank_indices)
drop_indices.update(blank_indices + 1)

# Drop the rows with these indices

dialogue = dialogue.drop(index=drop_indices)
dialogue = dialogue.reset_index(drop=True)

# Ensure both DataFrames have the same number of rows

if len(characters) != len(dialogue):
    raise ValueError("Both DataFrames must have the same number of rows")

# Insert the column from df2 into df1 at the second position

characters.insert(1, 'Dialogue', dialogue['Dialogue'])

# Insert a column from of the Eoisode name at the beginning

characters.insert(0, 'Episode Name', 'Overcast')
overcast=characters
print(overcast.head(1))
overcast.tail(1)
```
